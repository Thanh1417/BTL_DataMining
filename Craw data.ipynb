```python
import requests
from bs4 import BeautifulSoup #phân tích cú pháp html

categories = { #từ điển
    "Thể thao": ["https://vnexpress.net/the-thao", "https://vnexpress.net/the-thao-p2", "https://vnexpress.net/the-thao-p3", "https://vnexpress.net/the-thao-p4", "https://vnexpress.net/the-thao-p5"],
    "Giáo dục": ["https://vnexpress.net/giao-duc", "https://vnexpress.net/giao-duc-p2", "https://vnexpress.net/giao-duc-p3", "https://vnexpress.net/giao-duc-p4", "https://vnexpress.net/giao-duc-p5"],
    "Công nghệ": ["https://vnexpress.net/so-hoa", "https://vnexpress.net/so-hoa-p2", "https://vnexpress.net/so-hoa-p3", "https://vnexpress.net/so-hoa-p4", "https://vnexpress.net/so-hoa-p5"],
    "Kinh doanh": ["https://vnexpress.net/kinh-doanh", "https://vnexpress.net/kinh-doanh-p2", "https://vnexpress.net/kinh-doanh-p3", "https://vnexpress.net/kinh-doanh-p4", "https://vnexpress.net/kinh-doanh-p5"],
    "Sức khỏe": ["https://vnexpress.net/suc-khoe", "https://vnexpress.net/suc-khoe-p2", "https://vnexpress.net/suc-khoe-p3", "https://vnexpress.net/suc-khoe-p4", "https://vnexpress.net/suc-khoe-p5"],
    "Giải trí": ["https://vnexpress.net/giai-tri", "https://vnexpress.net/giai-tri-p2", "https://vnexpress.net/giai-tri-p3", "https://vnexpress.net/giai-tri-p4", "https://vnexpress.net/giai-tri-p5"]
}

data = []
headers = {"User-Agent": "Mozilla/5.0"}

for category, urls in categories.items():
    for url in urls:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            titles = soup.find_all(["h2", "h3"], class_=["title-news", "article-title"])

            for title in titles:
                title_text = title.text.strip()
                data.append({"Title": title_text, "Category": category})
        else:
            print(f"Lỗi {response.status_code} khi truy cập {url}")

print(f"Tổng số bài báo thu thập: {len(data)}")


# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(data)

# Kiểm tra dữ liệu trước khi xáo trộn
print("Dữ liệu trước khi xáo trộn:")
print(df.head(10))  # In 10 dòng đầu để kiểm tra

# Xáo trộn thứ tự dòng
df_shuffled = df.sample(frac=1, random_state=np.random.randint(0, 10000)).reset_index(drop=True)

# Kiểm tra dữ liệu sau khi xáo trộn
print("\nDữ liệu sau khi xáo trộn:")
print(df_shuffled.head(10))  # In lại 10 dòng đầu để kiểm tra

# Lưu file mới
output_path = "shuffled_data_vnexpress_titles.csv"
df_shuffled.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"\nDữ liệu đã được xáo trộn và lưu vào {output_path}")

from google.colab import files
files.download("shuffled_data_vnexpress_titles.csv")
```
